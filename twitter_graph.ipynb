```python
import json
with open('C:/Users/Kalyani/Documents/WebSearch_TextMining/project_WS/approach2_data.json', 'r') as fp:
    graph_data = json.load(fp)
```

```python
#handling plots
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nodes = []
#level1_level2_relationship
edges=[]
count = 0
nodes.append('tesla')
for i,j in graph_data['tesla'].items():
    nodes.append(i)
    tup_l1 = ("tesla",i)
    edges.append(tup_l1)
    if isinstance(j, list) and len(j) > 0:
        for k in j:
            if isinstance(j, list) and len(j) > 0:
                for l in k:
                    count+= 1
                    tup = (i,str(l))
                    nodes.append(str(l))
                    edges.append(tup)
            else:
                count+= 1
                nodes.append(str(k))
                tup = (i,str(k))
                edges.append(tup)
print(count)              
```

```python
import csv
with open('edges.csv', 'w') as outfile:
    csvwriter = csv.writer(outfile)
    csvwriter.writerows(edges) 
```

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
 
G=nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
 
```

```python
import random
node_filter = dict(G.degree(nodes))
nodes_plt = {}
count +=1 
edges_plt = {}
for i,j in node_filter.items():
    if j > 1:
        nodes_plt[i] = j
        count += j
        
nodes_final = random.choices(list(nodes_plt.keys()), k = 100)
print(nodes_final) 
# nodes_final 
```

```python
edges_list = []
for i in edges:
    if (i[0] in nodes_final) or (i[1] in nodes_final):
        if i[0] == 'tesla':
            tup = (i[0],i[1],10)
        else:
            tup = (i[0],i[1],5)
        edges_list.append(tup)
```

```python
import csv
with open('edges_filtered.csv', 'w') as outfile:
    csvwriter = csv.writer(outfile)
    csvwriter.writerows(edges_list)
```
